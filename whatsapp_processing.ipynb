```python
#https://towardsdatascience.com/complete-beginners-guide-to-processing-whatsapp-data-with-python-781c156b5f0b
import os

def read_file(file):
    '''Reads Whatsapp text file into a list of strings'''
    x = open(file,'r', encoding = 'utf-8') #Opens the text file into variable x but the variable cannot be explored yet
    y = x.read() #By now it becomes a huge chunk of string that we need to separate line by line
    content = y.splitlines() #The splitline method converts the chunk of string into a list of strings
    return content

chat = read_file('WhatsApp Chat with NombreGrupo.txt')
print(len(chat))
chat
```

```python
import datetime
# Remove elements that are not date

for i in range(len(chat)):
    try:
        datetime.datetime.strptime(chat[i].split(',')[0], '%x') #Converts string date into a date object
    except ValueError: #Returns an error if the string is not a datetime object
        chat[i-1] = chat[i-1] + ' ' + chat[i] #Appends the next line to the previous line
        chat[i] = "NA" #Replace the unwanted text element with 'NA'
    
#Handle more than double-line texting
for i in range(len(chat)):
      if chat[i].split(' ')[0] == 'NA':
        chat[i] = 'NA'
    
while True:
    try:
        chat.remove("NA")
    except ValueError:
        break
        
print(len(chat))
chat
```

```python
#date = [chat[i].split(',')[0] for i in range(len(chat))]
date = [text.split(',')[0] for text in chat]
```

```python
## Get time
time = [chat[i].split(',')[1].split('-')[0] for i in range(len(chat))]
time = [s.strip(' ') for s in time] # Remove spacing  
```

```python
## Get name
name = [chat[i].split('-')[1].split(':')[0] for i in range(len(chat))]
```

```python
content = []
for i in range(len(chat)):
    try:
        content.append(chat[i].split(':')[2])
    except IndexError:
        content.append('Missing Text')
```

```python
def find_feature(df,nombre,texto):
    lista=[]
    for i in df.Content:
        lista.append(texto in i)
        #print(texto in i)
    df[nombre]=lista
    df[nombre]=df[nombre].apply(lambda x: 1 if x == True else 0)
    return(df)
```

```python
feature1=[
'sign',
'entrar',
]
feature2=[
'plata',
'pagos',
'pagar'
]
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(date)), columns = ['Date'])

lista=[]
for i in df.Date:
    lista.append(datetime.datetime.strptime(i, '%x').date().year)
df['Date_Year']=lista

lista=[]
for i in df.Date:
    lista.append(datetime.datetime.strptime(i, '%x').date().month)
df['Date_Month']=lista

lista=[]
for i in df.Date:
    lista.append(datetime.datetime.strptime(i, '%x').date().day)
df['Date_Day']=lista
df['Time']=time
df['Name']=name
df['Content']=content
```

```python
df
```

```python
import gensim

data=df.Content.values.tolist()

def token_clean(text):
    for sentence in text:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=False))  # deacc=True removes punctuations

data_words = list(token_clean(data))

print(data_words[0])
```

```python
lista=[]
for i in data_words:
    lista.append(' '.join(i))
lista

df['Content']=lista
```

```python
df
```

```python
for f in feature1:
    find_feature(df,f,f)
```

```python
for f in feature2:
    find_feature(df,f,f)
```
